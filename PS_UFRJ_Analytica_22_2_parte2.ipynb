#4. Iteração em um Numpy array
Para iterar nos elementos costumamos utilizar o loop da mesma forma que uma
lista.
A partir do seguinte array: 
[5.5, 3.8, 9, 7.5, 10.0, 9.9, 8.5]

(a) Calcule a média dos valores usando iteração.

(b) Calcule o desvio padrão dos valores usando iteração.
Referência: https://www.todamateria.com.br/desvio-padrao/

```python
v=[5.5, 3.8, 9, 7.5, 10.0, 9.9, 8.5]
s=D=0
for i in range(len(v)):
  s+=v[i]
m=s/(i+1)
print('a)',m)
for i in range(len(v)):
  D+=(v[i]-m)**2
DP= (D/(i+1))**1/2
print('b)',DP)
```
